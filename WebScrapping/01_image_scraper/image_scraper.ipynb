```python
image_tags =soup.find_all("img")
```

```python
del image_tags[0:8]
```

```python
for i in image_tags:
    print(i['src2'])        
```

```python
save_dir = "images/"
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
for i in image_tags:
    image_data = requests.get(i['src2']).content
    with open(os.path.join(save_dir, f"{query}_{image_tags.index(i)}.jpg"), "wb") as f:
        f.write(image_data)
```
