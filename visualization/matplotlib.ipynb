Matplot is a plotting library for the Python programming language and its numerical mathematics extenson NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, QT, or GTK+.

Some of the major Pros of Matplotlib are:

1. Generally easy to get started for simple plots
2. Support for custom labels and texts
3. Great control of every element in a figure
4. High-quality output in many formats
5. Very customizable in general

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange(0, 10)
y = np.arange(11, 21)
```

```python
x
```

```python
y
```

```python
plt.scatter(x, y, c='green')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('X & y plot')
plt.savefig('test.jpg')
```

```python
plt.plot(x,y, c='yellow')
plt.xlabel('x Axis')
plt.ylabel('y Axis')
plt.title('Line plot')
```

```python
y = x*x
```

```python
plt.plot(x, y, 'r*')
plt.xlabel('x Axis')
plt.ylabel('y Axis')
plt.title('Line plot')
```

```python
plt.plot(x,y,'go')
```

```python
plt.plot(x, y, 'b+', linestyle='dashed', linewidth=3, markersize=10)
```

```python
plt.plot(x, y, 'y*', linestyle="dashed", linewidth=2.0, markersize=10)
```

```python
# Creating subplots
plt.subplot(2,2,1)
plt.plot(x, y, 'r--')
plt.xlabel('X axis')
plt.subplot(2,2,2)
plt.plot(y,x,'g--')
plt.xlabel('x axis 2')
plt.subplot(2,2,3)
plt.plot(y,x,'r*')
plt.subplot(2,2,4)
plt.plot(x, y, 'b+')
```

```python
# bar plots

x = [2,3,4,5]
y = [11,17,19,30]

x2 = [3,9,12,19]
y2 = [3, 12, 17, 32]

plt.bar(x, y, color='orange')
plt.bar(x2, y2)
```

```python
plt.bar(x2, y2, color='yellow')
```

```python
plt.bar(x,y, color='red')
```

```python
# Histogram
a = np.array([22, 4, 4, 23, 85, 76, 56, 9, 78, 66, 87, 56, 44])
plt.hist(a, bins=5)
```

```python
# Boxplot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
data
```

```python
plt.boxplot(data)
plt.show()
```

```python
# Pie chart
labels = 'Python', 'Java', 'R', 'SQL'
sizes = [215, 130, 245, 210]
colors = ['gold', 'lightcoral', 'yellowgreen', 'lightskyblue']
explode = (0.2, 0, 0, 0)

#plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=False)

plt.axis('equal')
plt.show()
```
